```python
TOPIC = "canzoni"
SERVER = 'localhost:9092'


import numpy as np
from json import dumps
import pandas as pd
import pip
from kafka import KafkaProducer
import time
import random



#con KafkaProducer creiamo il nostro object prodocer
if __name__ == "__main__":

    producer = KafkaProducer(bootstrap_servers=SERVER, value_serializer=lambda x: x.encode('utf-8'))
    
    #recuperiamo il dataset contenente le canzoni di artisti emergenti
    csv = "every_fresh_find.csv"
    canzoni = pd.read_csv(csv)
    canzoni['order_id'] = np.arange(len(canzoni))
    
    canzoni['Artist Name(s)'] = canzoni['Artist Name(s)'].str.replace('[^a-zA-Z]', '')
    canzoni['Artist IDs'] = canzoni['Artist IDs'].str.replace('[^a-zA-Z]', '')
    
    
    
    
    lista = canzoni.to_dict(orient="records")
    send = []    
    send = None

    for track  in lista:
        
        sending = []
        sending.append(track["order_id"])
        sending.append(track["Spotify ID"])
        sending.append(track["Track Name"])
        sending.append(track["Popularity"])
        sending.append(track["Duration (ms)"])
        sending.append(track["Artist Name(s)"])
        sending.append(track["Artist IDs"])
        sending.append(track["Release Date"])
        sending.append(track["Danceability"])
        sending.append(track["Energy"])
        sending.append(track["Key"])
        sending.append(track["Loudness"])
        sending.append(track["Mode"])
        sending.append(track["Speechiness"])
        sending.append(track["Acousticness"])
        sending.append(track["Instrumentalness"])
        sending.append(track["Liveness"])
        sending.append(track["Valence"])
        sending.append(track["Tempo"])
        sending.append(track["Time Signature"])
        
        track = ','.join(str(v) for v in sending)

        
        print("prossima canzone:" )
        print("song: ", track)
        producer.send(TOPIC, track)
        time.sleep(1)


    print("TOPIC COMPLETATO")
```
